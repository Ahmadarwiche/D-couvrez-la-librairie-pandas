```python
liste_zone = ['eastern_africa', 'middle_africa', 'northern_africa', 'southern_africa', 'western_africa', 'northern_america', 'central_america', 'caribbean', 'south_america', 'central_asia', 'eastern_asia', 'southern_asia', 'south_eastern_asia', 'western_asia', 'eastern_europe', 'northern_europe', 'southern_europe', 'western_europe', 'australia_new_zealand', 'melanesia', 'micronesia', 'polynesia'
]
```

```python
def Data_pays_zone(zone):
    df_z = pd.read_csv(zone+'.csv')
    df_z['zone']=str(zone)
    return df_z[['Area','zone']]

df_concat= pd.concat([Data_pays_zone(zone) for zone in liste_zone])
df_concat
```

```python
df_concat = df_concat.rename(columns=lambda x : x.strip().lower().replace(' ', '_'))
```

```python
df = df.merge(df_concat, on=['area'])
df
```
